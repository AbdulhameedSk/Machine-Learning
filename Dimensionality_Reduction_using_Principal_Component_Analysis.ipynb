```python
import pandas as pd
import matplotlib.pyplot as plt
from sklearn.decomposition import PCA
from sklearn.datasets import load_breast_cancer


from sklearn.decomposition import PCA
from sklearn.datasets import load_breast_cancer
from sklearn.preprocessing import StandardScaler


from sklearn.preprocessing import StandardScaler
```

```python
df=load_breast_cancer()
```

```python
features=df.data
target=df.target
```

```python
#standardize the data
scalar=StandardScaler()
scaled_data=scalar.fit_transform(features)
```

```python
# Perform PCA
pca = PCA(n_components=2)  # Set the desired number of components
principal_components = pca.fit_transform(scaled_data)
```

```python
# Access the principal components
principal_component_1 = principal_components[:, 0]
principal_component_2 = principal_components[:, 1]
```

```python
plt.scatter(principal_component_1,principal_component_2,c=target)
plt.colorbar(label="Target")
plt.show()
```
